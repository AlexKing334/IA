## Step 2: Create a CNN to Classify Dog Breeds (from scratch)

```python
# Imports
import torch
from torchvision.datasets import ImageFolder
import torchvision.transforms as T
from torch.utils.data import DataLoader

# Data loading options
batch_size = 8
load_size = 200
crop_size = 200
norm_mean = (0.5, 0.5, 0.5)
norm_std = (0.5, 0.5, 0.5)

# Train transform
train_transform = T.Compose([
    T.Resize(load_size),
    T.RandomCrop(crop_size),
    T.RandomHorizontalFlip(),
    T.ToTensor(),
    T.Normalize(mean=norm_mean,std=norm_std)
])

# Test transform
test_transform = T.Compose([
    T.Resize(load_size),
    T.CenterCrop(crop_size),
    T.ToTensor(),
    T.Normalize(mean=norm_mean,std=norm_std) 
])

# Instantiate datasets
dog_train_dataset = ImageFolder(os.path.join(root_dir, "dog_images", "train"), transform=train_transform)
dog_val_dataset = ImageFolder(os.path.join(root_dir, "dog_images", "valid"), transform=test_transform)
dog_test_dataset = ImageFolder(os.path.join(root_dir, "dog_images", "test"), transform=test_transform)

# Get number of classes (we'll need it in the model)
num_classes = len(dog_train_dataset.classes)

# Print dataset statistics
print(f"Num. classes: {num_classes}")
print(f"Num. train samples: {len(dog_train_dataset)}")
print(f"Num. valid. samples: {len(dog_val_dataset)}")
print(f"Num. test samples: {len(dog_test_dataset)}")


# Instantiate data loaders
train_loader = DataLoader(dog_train_dataset, batch_size=batch_size, num_workers=1, shuffle=True)
val_loader   = DataLoader(dog_val_dataset,   batch_size=batch_size, num_workers=1, shuffle=False)
test_loader  = DataLoader(dog_test_dataset,  batch_size=batch_size, num_workers=1, shuffle=False)

loaders = {"train": train_loader,
           "val": val_loader,
           "test": test_loader}
```

```python
# Imports
import torch.nn as nn
import torch.nn.functional as F
class Model_tmp(nn.Module):
  def __init__(self):
        # Call parent constructor
        super().__init__();
        # Create convolutional layers
        self.conv_layers = nn.Sequential(
            # Layer 1
            nn.Conv2d(3, 64, kernel_size=3, padding=0, stride=1),
            nn.ReLU(),
            nn.BatchNorm2d(64),
            # Layer 2
            nn.Conv2d(64, 128, kernel_size=3, padding=0, stride=1),
            nn.ReLU(),
            nn.BatchNorm2d(128),
            # Layer 3
            nn.Conv2d(128, 128, kernel_size=3, padding=0, stride=1),
            nn.ReLU(),
            nn.BatchNorm2d(128),
            nn.MaxPool2d(kernel_size=2, stride=2),
            # Layer 4
            nn.Conv2d(128, 256, kernel_size=3, padding=0, stride=1),
            nn.ReLU(),
            nn.BatchNorm2d(256),
            nn.MaxPool2d(kernel_size=2, stride=2)
          
           
        )
  def forward(self, x):
        return self.conv_layers(x)
```

```python
# Test model output
modelTmp = Model_tmp()
test_input = dog_train_dataset[0][0].unsqueeze(0)
output_size=modelTmp(test_input).size()
input_size=output_size[1]*output_size[2]*output_size[3]
```

```python
# Imports
import torch.nn as nn
import torch.nn.functional as F

# Define model
class Model(nn.Module):
# Constructor
    def __init__(self):
        # Call parent constructor
        super().__init__();
        # Create convolutional layers
        self.conv_layers = nn.Sequential(
            # Layer 1
            nn.Conv2d(3, 64, kernel_size=3, padding=0, stride=1),
            nn.ReLU(),
            nn.BatchNorm2d(64),
            # Layer 2
            nn.Conv2d(64, 128, kernel_size=3, padding=0, stride=1),
            nn.ReLU(),
            nn.BatchNorm2d(128),
            # Layer 3
            nn.Conv2d(128, 128, kernel_size=3, padding=0, stride=1),
            nn.ReLU(),
            nn.BatchNorm2d(128),
            nn.MaxPool2d(kernel_size=2, stride=2),
            # Layer 4
            nn.Conv2d(128, 256, kernel_size=3, padding=0, stride=1),
            nn.ReLU(),
            nn.BatchNorm2d(256),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        # Create fully-connected layers
        self.fc_layers = nn.Sequential(
            # FC layer
            nn.Linear(input_size, 1024),
            nn.Dropout(0.5),
            nn.ReLU(),
            # Classification layer
            nn.Linear(1024, num_classes)
        )

    # Forward
    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)
        x = self.fc_layers(x)
        return x
```

```python
# Create model
model = Model()
```

```python
# Test model output
model.eval()
test_input = dog_train_dataset[0][0].unsqueeze(0)
print("Model output size:", model(test_input).size())
```

```python
# Move model to device
model.to(dev)
```

```python
# Create optimizer
import torch.optim as optim
optimizer = optim.SGD(model.parameters(), lr=0.001)
```

```python
# Initialize training history
# HINT: here you set up the lists for average loss/accuracy for all epochs
history_loss = {"train": [], "val": [], "test": []}
history_accuracy = {"train": [], "val": [], "test": []}
# Initialize best validation accuracy and test accuracy at best validation accuracy
best_val_accuracy = 0
test_accuracy_at_best_val = 0
```

```python
# Start training
# HINT:
# - change the number of epochs as you want
# - you can always interrupt training by clicking on the "interrupt kernel" on the toolbar
# - you can set a small number of epochs, train for a while, plot losses, and then train again
# - in general, training for at most 50 epochs will be ok
try:
  for epoch in range(50):
    # Initialize variables for computing average loss/accuracy
    sum_loss = {"train": 0, "val": 0, "test": 0}
    sum_accuracy = {"train": 0, "val": 0, "test": 0}
    # Process each split
    for split in ["train", "val", "test"]:
        # Set network mode (train or eval)
      if split=="train":
          model.train()
      else:
          model.eval()
        # Process all data in split
      for input, labels in loaders[split]:
            # Move to CUDA
        input = input.to(dev)
        labels = labels.to(dev)
        optimizer.zero_grad()
            # Compute loss
        pred = model(input)
        loss = F.cross_entropy(pred, labels)
            # Update variables for average epoch loss
        sum_loss[split] += loss.item()
            # Compute accuracy
        _,pred_labels = pred.max(1)
        batch_accuracy = (pred_labels == labels).sum().item()/input.size(0)
            # Update variables for average epoch accuracy
        sum_accuracy[split] += batch_accuracy
            # Backward and optimize
        if split == "train":
          loss.backward() 
          optimizer.step()
    # Compute average epoch loss/accuracy
    epoch_loss = {split: sum_loss[split]/len(loaders[split]) for split in ["train", "val", "test"]}
    epoch_accuracy = {split: sum_accuracy[split]/len(loaders[split]) for split in ["train", "val", "test"]}
    # Append average values to history lists
    for split in ["test", "val", "train"]:
      history_loss[split].append(epoch_loss[split])
      history_accuracy[split].append(epoch_accuracy[split])
      
    # Check if we obtained the best validation accuracy
      if split == "val":
          temp_max_value = max(history_accuracy[split]) 
          if history_accuracy[split][-1] == temp_max_value:
                # Update best validation accuracy and test accuracy at best validation
             best_val_accuracy = temp_max_value
             test_accuracy_at_best_val = history_accuracy["test"][epoch]
    
     # Print info
    print(f"Epoch {epoch+1}:",
                  f"TrL={epoch_loss['train']:.4f},",
                  f"TrA={epoch_accuracy['train']:.4f},",
                  f"VL={epoch_loss['val']:.4f},",
                  f"VA={epoch_accuracy['val']:.4f},",
                  f"TeL={epoch_loss['test']:.4f},",
                  f"TeA={epoch_accuracy['test']:.4f},")
except KeyboardInterrupt:
        print("Interrupted")     
```

```python
# Data loading options
batch_size = 8
load_size = 224
crop_size = 224
norm_mean = (0.485, 0.456, 0.406)
norm_std = (0.229, 0.224, 0.225)

# Train transform
train_transform = T.Compose([
    T.Resize(load_size),
    T.RandomCrop(crop_size),
    T.RandomHorizontalFlip(),
    T.ToTensor(),
    T.Normalize(mean=norm_mean,std=norm_std)
])

# Test transform
test_transform = T.Compose([
    T.Resize(load_size),
    T.CenterCrop(crop_size),
    T.ToTensor(),
    T.Normalize(mean=norm_mean,std=norm_std)
])

# Instantiate datasets
dog_train_dataset = ImageFolder(os.path.join(root_dir, "dog_images", "train"), transform=train_transform)
dog_val_dataset = ImageFolder(os.path.join(root_dir, "dog_images", "valid"), transform=test_transform)
dog_test_dataset = ImageFolder(os.path.join(root_dir, "dog_images", "test"), transform=test_transform)

# Get number of classes (we'll need it in the model)
num_classes = len(dog_train_dataset.classes)

# Print dataset statistics
print(f"Num. classes: {num_classes}")
print(f"Num. train samples: {len(dog_train_dataset)}")
print(f"Num. valid. samples: {len(dog_val_dataset)}")
print(f"Num. test samples: {len(dog_test_dataset)}")


# Instantiate data loaders
train_loader = DataLoader(dog_train_dataset, batch_size=batch_size, num_workers=1, shuffle=True)
val_loader   = DataLoader(dog_val_dataset,   batch_size=batch_size, num_workers=1, shuffle=False)
test_loader  = DataLoader(dog_test_dataset,  batch_size=batch_size, num_workers=1, shuffle=False)

loaders = {"train": train_loader,
           "val": val_loader,
           "test": test_loader}
```

```python
class AlexNetTmp(nn.Module):
    def __init__(self):
        super().__init__()
        # Load AlexNet model
        alexnet = models.alexnet(pretrained=True)
        # Select feature extraction part
        self.features = alexnet.features    
    
    def forward(self, x):
        # Go through the pre-trained layers
        x = self.features(x)
        return x
```

```python
# Test model output
alexnetTmp = AlexNetTmp()
test_input = dog_train_dataset[0][0].unsqueeze(0)
output_size=alexnetTmp(test_input).size()
input_size=output_size[1]*output_size[2]*output_size[3]
```

```python
print("Output Features:",input_size)
```

```python
# Define fine-tuned model
class FineTunedAlexNet(nn.Module):    
    def __init__(self):
        super().__init__()
        # Load AlexNet model
        alexnet = models.alexnet(pretrained=True)
        # Select feature extraction part
        self.features = alexnet.features
        self.fc_layers = nn.Sequential(
            # FC layer
            nn.Linear(input_size, 1024),
            nn.Dropout(0.5),
            nn.ReLU(),
            # Classification layer
            nn.Linear(1024, num_classes)
        )
   
    def forward(self, x):
        # Go through the pre-trained layers
        x = self.features(x)
        x = x.view(x.size(0), -1)
        x = self.fc_layers(x)
        return x
```

```python
#Create model
model = FineTunedAlexNet()
```

```python
# Test model output
model.eval()
test_input = dog_train_dataset[0][0].unsqueeze(0)
print("Model output size:", model(test_input).size())
```

```python
# Move model to CUDA
model.to(dev)
```

```python
#Create optimizer
optimizer = optim.SGD(model.parameters(), lr=0.001)
```

```python
# Initialize training history
history_loss = {"train": [], "val": [], "test": []}
history_accuracy = {"train": [], "val": [], "test": []}
# Initialize best validation accuracy and test accuracy at best validation accuracy
best_val_accuracy = 0
test_accuracy_at_best_val = 0
```

```python
# Start training
# HINT:
# - change the number of epochs as you want
# - you can always interrupt training by clicking on the "interrupt kernel" on the toolbar
# - you can set a small number of epochs, train for a while, plot losses, and then train again
# - in general, training for at most 50 epochs will be ok
try:
  for epoch in range(50):
    # Initialize variables for computing average loss/accuracy
    sum_loss = {"train": 0, "val": 0, "test": 0}
    sum_accuracy = {"train": 0, "val": 0, "test": 0}
    # Process each split
    for split in ["train", "val", "test"]:
        # Set network mode (train or eval)
      if split=="train":
          model.train()
      else:
          model.eval()
        # Process all data in split
      for input, labels in loaders[split]:
            # Move to CUDA
        input = input.to(dev)
        labels = labels.to(dev)
        optimizer.zero_grad()
            # Compute loss
        pred = model(input)
        loss = F.cross_entropy(pred, labels)
            # Update variables for average epoch loss
        sum_loss[split] += loss.item()
            # Compute accuracy
        _,pred_labels = pred.max(1)
        batch_accuracy = (pred_labels == labels).sum().item()/input.size(0)
            # Update variables for average epoch accuracy
        sum_accuracy[split] += batch_accuracy
            # Backward and optimize
        if split == "train":
          loss.backward() 
          optimizer.step()
    # Compute average epoch loss/accuracy
    epoch_loss = {split: sum_loss[split]/len(loaders[split]) for split in ["train", "val", "test"]}
    epoch_accuracy = {split: sum_accuracy[split]/len(loaders[split]) for split in ["train", "val", "test"]}
    # Append average values to history lists
    for split in ["test", "val", "train"]:
      history_loss[split].append(epoch_loss[split])
      history_accuracy[split].append(epoch_accuracy[split])
      
    # Check if we obtained the best validation accuracy
      if split == "val":
          temp_max_value = max(history_accuracy[split]) 
          if history_accuracy[split][-1] == temp_max_value:
                # Update best validation accuracy and test accuracy at best validation
             best_val_accuracy = temp_max_value
             test_accuracy_at_best_val = history_accuracy["test"][epoch]
    
     # Print info
    print(f"Epoch {epoch+1}:",
                  f"TrL={epoch_loss['train']:.4f},",
                  f"TrA={epoch_accuracy['train']:.4f},",
                  f"VL={epoch_loss['val']:.4f},",
                  f"VA={epoch_accuracy['val']:.4f},",
                  f"TeL={epoch_loss['test']:.4f},",
                  f"TeA={epoch_accuracy['test']:.4f},")
except KeyboardInterrupt:
        print("Interrupted")
```
